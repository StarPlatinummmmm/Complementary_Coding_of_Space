```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the first file
data1 = np.load('bump_scores.npz')
bump_score_diff_1 = data1['bump_score_diff']
bump_score_diff_std_1 = data1['bump_score_diff_std']

# Load data from the second file
data2 = np.load('bump_scores_coupled_net.npz')
bump_score_diff_2 = data2['bump_score_diff']
bump_score_diff_std_2 = data2['bump_score_diff_std']

# Plot error bars
x = np.arange(len(bump_score_diff_1))+2  # Assuming the scores are 1-dimensional

plt.figure(figsize=(6, 5))

# Plot data from the first file
plt.plot(x, bump_score_diff_1, label='940 Place cells')
plt.plot(x, bump_score_diff_2, label='800 Place cells with 20*7 grid cells')
plt.legend()
plt.errorbar(x, bump_score_diff_1, yerr=bump_score_diff_std_1, fmt='o', capsize=5)
plt.errorbar(x, bump_score_diff_2, yerr=bump_score_diff_std_2, fmt='s', capsize=5)

plt.xlabel('Embedded Map Number')
plt.ylabel('Bump Score Difference')
plt.title('Errorbar Plot of Bump Score Differences')
plt.legend()
plt.grid(True)

plt.show()
```
