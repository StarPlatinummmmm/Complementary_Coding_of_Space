```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load the Excel file
file_path = "neurons_data.xlsx"  # 请将此路径替换为实际文件路径
df = pd.read_excel(file_path)

# Extract the three matrices
neuron_number = df["neuron number"].values
only_place_cells = df.iloc[:, 1:8].values
coupled_net = df.iloc[:, 8:14].values

# print("Neuron Number:", neuron_number)
# print("Only Place Cells:")
# print(only_place_cells)
# print("Coupled Net:")
# print(coupled_net)

map_num_only = only_place_cells
mean_map_num_only = np.mean(map_num_only,axis=1)
std_map_num_only = np.std(map_num_only,axis=1)
mean_map_num_couple = np.mean(coupled_net,axis=1)
std_map_num_couple = np.std(coupled_net,axis=1)

plt.rcParams.update({'font.size': 15, 'font.family': 'Arial'})
plt.figure(figsize=(4,4))
plt.plot(neuron_number, mean_map_num_only, 'b', label='only place cells')
plt.plot(neuron_number, mean_map_num_couple, 'r', label='place cells x grid cells')
plt.legend()
plt.errorbar(neuron_number, mean_map_num_only, yerr=std_map_num_only, fmt='bo', capsize=5, label='only place cells')
plt.errorbar(neuron_number, mean_map_num_couple, yerr=std_map_num_couple, fmt='rs', capsize=5, label='place cells x grid cells')
plt.xlabel('Total neuron number')
plt.ylabel('Stored Map number')
plt.title('Memory capacity')
plt.tight_layout()
plt.savefig('Stored_map_number.pdf')
```
