```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 设置横轴范围
x = np.linspace(-50, 50, 1000)

# 创建一个图像
fig, ax = plt.subplots()

# 定义小圈的横坐标范围，每隔2画一个
circle_x = np.arange(-15, 16, 2)

# 定义高斯波包函数
def gaussian(x, mu, sigma=3.0):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

# 设置颜色渐变
colors = cm.viridis(np.linspace(0, 1, len(circle_x)))

# 画小圈和高斯波包
for i, cx in enumerate(circle_x):
    ax.plot(cx, -0.1, 'o', color=colors[i], markersize=5)  # 画小圈，位于横线下方
    gaussian_y = gaussian(x, cx)  # 高斯波包
    ax.plot(x, gaussian_y, color=colors[i])  # 画高斯波包

# 设置图像范围
ax.set_xlim([-50, 50])
ax.set_ylim([-1, 1.5])

# 移除横纵坐标
ax.axis('off')

# 在图像下方添加 "position" 标签
plt.text(50, -0.2, 'Position', ha='center', va='center')

# 画黑色横线并加上向右箭头
ax.plot(x, np.zeros_like(x), color='black', linewidth = 2)
ax.annotate('', xy=(50, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", color='black'))

# 显示图像
# 保存图像为矢量图 (SVG) 格式，去掉白色背景
fig.patch.set_alpha(0)
plt.savefig("place_cell_coding.svg", format="svg", transparent=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 设置横轴范围
x = np.linspace(-50, 50, 2000)

# 创建一个图像
fig, ax = plt.subplots()

# 定义小圈的横坐标范围，每隔2画一个
circle_x = np.arange(-15, 16, 2)

# 定义高斯波包函数
def gaussian(x, mu, sigma=3.0):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

# 设置第一个pattern的颜色渐变
colors_viridis = cm.viridis(np.linspace(0, 1, len(circle_x)))

# 画第一个pattern
for i, cx in enumerate(circle_x):
    ax.plot(cx, -0.1, 'o', color=colors_viridis[i], markersize=5)  # 画小圈，位于横线下方
    gaussian_y = gaussian(x, cx)  # 高斯波包
    ax.plot(x, gaussian_y, color=colors_viridis[i])  # 画高斯波包

# 设置第二个pattern的颜色渐变
colors_plasma = cm.plasma(np.linspace(0, 1, len(circle_x)))

# 画第二个pattern
for i, cx in enumerate(circle_x):
    cx_shifted = cx + 30  # 第二个pattern向右平移
    ax.plot(cx_shifted, -0.1, 'o', color=colors_plasma[i], markersize=5)  # 画小圈，位于横线下方
    gaussian_y = gaussian(x, cx_shifted)  # 高斯波包
    ax.plot(x, gaussian_y, color=colors_plasma[i])  # 画高斯波包

# 设置图像范围
ax.set_xlim([-50, 100])
ax.set_ylim([-1, 1.5])

# 移除横纵坐标
ax.axis('off')

# 画黑色横线并加上向右箭头
ax.plot(np.linspace(-50, 100, 1500), np.zeros(1500), color='black', linewidth=2)
ax.annotate('', xy=(100, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", color='black'))

# 在图像下方添加 "position" 标签
plt.text(100, -0.2, 'Position', ha='center', va='center')

# 保存图像为矢量图 (SVG) 格式，去掉白色背景
fig.patch.set_alpha(0)
plt.savefig("place_cell_coding.svg", format="svg", transparent=True)

# 显示图像
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 设置横轴范围
x = np.linspace(-50, 50, 3000)  # 横轴范围扩大，适应多次重复

# 创建一个图像
fig, ax = plt.subplots()

# 定义小圈的横坐标范围，每隔2画一个
circle_x = np.arange(-15, 15, 2)

# 定义高斯波包函数
def gaussian(x, mu, sigma=3.0):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

# 设置颜色渐变
colors = cm.viridis(np.linspace(0, 1, len(circle_x)))

# 画重复的pattern
for shift in [-30, 0, 30]:  # 左右延伸
    for i, cx in enumerate(circle_x):
        cx_shifted = cx + shift
        ax.plot(cx_shifted, -0.1, 'o', color=colors[i], markersize=5)  # 画小圈，位于横线下方
        gaussian_y = gaussian(x, cx_shifted)  # 高斯波包
        ax.plot(x, gaussian_y, color=colors[i])  # 画高斯波包

# 设置图像范围
ax.set_xlim([-50, 50])
ax.set_ylim([-1, 1.5])

# 移除横纵坐标
ax.axis('off')

# 画黑色横线并加上向右箭头
ax.plot(x, np.zeros_like(x), color='black', linewidth=2)
ax.annotate('', xy=(50, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", color='black'))

# 在图像下方添加 "position" 标签
plt.text(50, -0.2, 'Position', ha='center', va='center')

# 保存图像为矢量图 (SVG) 格式，去掉白色背景
fig.patch.set_alpha(0)
plt.savefig("pattern.svg", format="svg", transparent=True)

# 显示图像
plt.show()
```
